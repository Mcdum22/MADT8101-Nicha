# **Customer** **Single View**

### **What is a Customer Single View?**

A Customer Single View, also called a unified customer view,  is an accurate, consistent and holistic record of the data held by a company about its customers, and can be viewed in one place.

A Single Customer View collects 3 types of data:

**•	Demographic data**: includes profile data such as gender, age, address, household data, etc.

**•	Transactional data**: includes purchases, dates, abandoned carts, returns, product value, etc.

**•	Behavioral data**: includes data from web visits, mobile app and in-store, such as most-visited products and categories, number of pages/stores visited, interactions, preferred channels, lifestyle, etc.

### **Why Customer Single View is Importance?**

**No more data duplication**

Customer Single View resolve the pervasive problem of data redundancy within your organization. Duplicate data introduces a significant risk of errors and inconsistencies.

**A better understanding of every single customer**

Customer Single Views consolidate the myriad dispersed data fragments gathered throughout your organization. They construct a comprehensive roadmap that vividly depicts the holistic nature of customer interactions with your brand, their purchasing patterns, and their engagements with your customer service teams. CSVs empower you to:


*   Swiftly access every customer interaction, ensuring you enter conversations well-prepared and informed about ticket statuses.
  
*   Eliminate the hassle of navigating standalone service applications.


*   Access a multitude of data points within a unified platform, expediting,communication, enhancing customer experiences, and accelerating the resolution of customer issues.

```python
df['SHOP_DATE'] = pd.to_datetime(df['SHOP_DATE'], format='%Y%m%d')
df.head()
```

```python
df_mas = df[df['CUST_CODE'].str.contains("CUST",na = False)]
```

```python
df_mas.info()
```

```python
from statistics import mode

def get_mode(x):
    try:
        return mode(x)
    except:
        return None
```

```python
num_sku = df_mas.PROD_CODE.nunique()
```

```python
# df_mas['visit'] = df_mas['date'].dt.strftime('%Y-%m-%d') + '_' + df_mas['customer_id'].astype(str)
num_sku = df_mas.PROD_CODE.nunique()

agg_funcs = {
              'SPEND': ['sum','mean']
              , 'QUANTITY': ['sum','mean']
              , 'PROD_CODE': lambda x: len(x.unique())/num_sku

              , 'BASKET_ID': lambda x: len(x.unique())
              , 'SHOP_DATE': lambda x: len(x.unique())
              , 'PROD_CODE_20': lambda x: get_mode(x)
              , 'PROD_CODE_40': lambda x: get_mode(x)
              , 'BASKET_SIZE': lambda x: get_mode(x)
              , 'BASKET_PRICE_SENSITIVITY': lambda x: get_mode(x)
              , 'BASKET_TYPE': lambda x: get_mode(x)
              , 'BASKET_DOMINANT_MISSION': lambda x: get_mode(x)

}


df_group = df_mas.groupby('CUST_CODE').agg(agg_funcs)

df_group.columns = ['_'.join(col).strip() for col in df_group.columns.values]
df_group.columns = ['total_amount_dollar_sales', 'average_dollar_sales','total_qty', 'average_qty
                    , 'sku_penetrate', 'number_tnx', 'number_visit', 'most_dept'
                    , 'most_division','most_basket_size','most_basket_sens','most_basket_type','most_basket_dominant']

df_group.reset_index(inplace = True)

df_group['atv'] = df_group['total_amount_dollar_sales']/ df_group['number_tnx']
df_group['spend_per_visit'] = df_group['total_amount_dollar_sales']/df_group['number_visit']


df_group.head()
```
